# 885. Spiral Matrix III
https://leetcode.com/problems/spiral-matrix-iii/description/?envType=daily-question&envId=2024-08-08

```python
from typing import List


rows = 5
cols = 6
rStart = 1
cStart = 4
w = max(rows,cols) #弄成正方形
graph_list= []
for i in range(w):
    temp = []
    for j in range(w):
        temp.append(0)
    graph_list.append(temp)

ans_list = [[rStart,cStart]]
total_move = [i for i in range(1,cols+1)]
c=cStart
r=rStart
head_x_dir = 1
head_y_dir = 1
round = 0

while len(ans_list)<rStart*cStart:

    c_move=total_move[round]*2
    while c_move>0: #走col方向

        if c+head_x_dir < cStart:
            c += head_x_dir
            graph_list[c][r] = 1
            if r < rows and c < cols:
                ans_list.append([r,c])
                c_move-=1

        else: #撞牆
            if c <= cStart and r >= rStart: #原點右方

                if r + total_move[round] < cStart: # 原點向下找
                    r = rStart+total_move[round]
                    head_x_dir *= -1 #方向向左
                
                else: # 原點向左找
                    
                    c = cStart-total_move[round]
                    head_y_dir *= -1 #方向向左

            else:                           #原點左方
                if c + total_move[round] < cStart: # 原點向右找
                    c = cStart+total_move[round]
                    head_x_dir *= -1 #方向向左
                else: # 原點向左找
                    c = cStart-total_move[round]
                    head_y_dir *= -1 #方向向左


    for _y in range(total_move[round]):
        if r+head_y_dir < cStart:
            r += head_y_dir    
            graph_list[c][r] = 1
            


  


```

```python
from typing import List

def spiralOrder(rows: int, cols: int, rStart: int, cStart: int) -> List[List[int]]:
    directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]  # Right, Down, Left, Up
    result = []
    r, c = rStart, cStart
    directionIndex = 0
    steps = 0  # Steps in the current direction
    change = 0  # Number of direction changes
    total_cells = rows * cols

    while len(result) < total_cells:
        print(steps,change // 2 + 1,[r, c],directions[directionIndex])
        if 0 <= r < rows and 0 <= c < cols:
            result.append([r, c])
        if steps == change // 2 + 1:
            
            directionIndex = (directionIndex + 1) % 4 # next direction %4 sliding window direction
            steps = 0
            change += 1
        r += directions[directionIndex][0]
        c += directions[directionIndex][1]
        steps += 1
    return result

# 测试例子
#print(spiralOrder(1, 4, 0, 0))  # Output: [[0,0],[0,1],[0,2],[0,3]]
print(spiralOrder(5, 6, 1, 4))  # Output: [[1,4],[1,5],[2,5],[2,4],[2,3],[1,3],[0,3],[0,4],[0,5],[3,5],[3,4],[3,3],[3,2],[2,2],[1,2],[0,2],[4,5],[4,4],[4,3],[4,2],[4,1],[3,1],[2,1],[1,1],[0,1],[4,0],[3,0],[2,0],[1,0],[0,0]]
```
