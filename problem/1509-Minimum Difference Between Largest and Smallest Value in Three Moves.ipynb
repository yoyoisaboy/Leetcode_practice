# 1509. Minimum Difference Between Largest and Smallest Value in Three Moves

```python
from typing import List
```

```python
def minDifference(nums: List[int]) -> int:
    if len(nums)<5:
        return 0
    nums.sort()
    ans = []
    left_loc = 0
    while left_loc!=4:
        slide_window = nums[left_loc:(len(nums)-3+left_loc)]
        #print(slide_window)
        ans.append( slide_window[-1]-slide_window[0] )
        left_loc +=1
    #print('ans:',ans)
    return min(ans)
```

## 方法 1：貪婪算法 (Greedy)
這個方法很直接：排序陣列，然後我們可以通過最多移除三個最小值或三個最大值來減少範圍。我們考慮四種情況，移除：

1. 三個最小值
2. 兩個最小值和一個最大值
3. 一個最小值和兩個最大值
4. 三個最大值

在這四種情況中取最小差值。
O(n log n)

```python
def minDifference(nums):
    if len(nums) <= 4:
        return 0
    
    nums.sort()
    return min(nums[-1] - nums[3], nums[-2] - nums[2], nums[-3] - nums[1], nums[-4] - nums[0])

# 測試用例
print(minDifference([5, 3, 2, 4]))  # 輸出：0
print(minDifference([1, 5, 0, 10, 14]))  # 輸出：1
print(minDifference([3, 100, 20]))  # 輸出：0
```

## 方法 2：堆積 (Heap)
堆積方法可以用來高效地找出最小值和最大值。具體來說，我們可以使用最大堆來找出最小的前四個數，最小堆來找出最大的前四個數。

堆積：使用最小堆和最大堆找出前四個數，時間複雜度為O(n log k)。

```python
import heapq

def minDifference(nums):
    if len(nums) <= 4:
        return 0
    
    # 最小的4個數
    min_nums = heapq.nsmallest(4, nums)
    # 最大的4個數
    max_nums = heapq.nlargest(4, nums)
    
    min_diff = float('inf')
    for i in range(4):
        min_diff = min(min_diff, max_nums[i] - min_nums[3 - i])
    
    return min_diff

# 測試用例
print(minDifference([5, 3, 2, 4]))  # 輸出：0
print(minDifference([1, 5, 0, 10, 14]))  # 輸出：1
print(minDifference([3, 100, 20]))  # 輸出：0
```
